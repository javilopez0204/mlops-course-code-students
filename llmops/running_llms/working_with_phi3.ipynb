```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_ollama.llms import OllamaLLM
from langchain.chains import ConversationChain
from langfuse.callback import CallbackHandler
from langchain.chains.conversation.memory import ConversationBufferMemory

class LangChainPipeline:
    def __init__(self, model_name: str, template: str, enable_tracing: bool = True):
        """
        Initializes the LangChain pipeline with a given model, prompt template, and memory.
        :param model_name: Name of the language model.
        :param template: Prompt template for generating responses.
        :param enable_tracing: Flag to enable or disable tracing.
        """
        self.model = OllamaLLM(model=model_name)
        self.prompt = ChatPromptTemplate.from_template(template)
        self.memory = ConversationBufferMemory(memory_key="chat_history")
        self.chain = ConversationChain(
                                        llm=self.model,
                                        memory=ConversationBufferMemory()
                                    )
        self.enable_tracing = enable_tracing
        if enable_tracing:
            self.langfuse_handler = CallbackHandler(
            public_key="pk-lf-867db27e-ff86-4374-83fa-7f04b9ed868b",
            secret_key="sk-lf-12290eb7-4fa3-4088-8b44-073135d29b6b",
            host="http://localhost:3000"
        )

    def invoke(self, question: str):
        """
        Invokes the chain with the given question while optionally tracing interactions and storing memory.
        :param question: User's question.
        :return: Model's response.
        """
        response = self.chain.predict(input=question, callbacks=[self.langfuse_handler])
        return response

template = """Question: {question}

Answer: """
model_name = "phi3:mini"
pipeline = LangChainPipeline(model_name, template, enable_tracing=True)
```

```python
response = pipeline.invoke("Do you remember my name?")
print(response)
```
